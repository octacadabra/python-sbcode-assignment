```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18,8), sharey=True)

ax[0].plot(x, tan, color='b', label='tanX')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, cos, color='r', label='sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, sin, color='g', label='cosX')
ax[2].set_title('Grafik cosx')

fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (16,8))
ax.plot (x,y)
ax.set_title('Grafik CosX')


ins4 = ax.inset_axes((0.88, 0.88, 0.3, 0.4))
ins5 = ax.inset_axes((-0.13,0.88,0.3,0.4))
ins2 = ax.inset_axes((-0.13,-0.13,0.3,0.4))
ins3 = ax.inset_axes((0.88,-0.13,0.3,0.4))

ins4.plot(x, y, 'b',  x, y4, 'black')
ins4.set_title('Grafik Cos^4')
ins5.plot(x, y, 'b', x, y5, 'y')
ins5.set_title('Grafik Cos^5')
ins2.plot(x, y, 'b', x, y2, 'r')
ins2.set_title('Grafik Cos^2')
ins3.plot(x, y, 'b', x, y3, 'g')
ins3.set_title('Grafik Cos^3')

plt.show()
```
