- Apa itu machine learning?

Machine Learning merupakan salah satu cabang dari kecerdasan buatan. Machine Learning mampu mempelajari data yang ada dan melakukan tugas-tugas tertentu sesuai dengan apa yang telah dipelajari. Machine Learning mampu melakukan hal tersebut dengan metode mempelajari algoritma dan model statistik yang ada. Machine Learning memiliki kemampuan untuk memperoleh data yang ada dengan perintahnya sendiri. Kemudian mempelajari data yang ada sehingga bisa melakukan tugas tertentu.

- Apa itu data feature dan data target?

Data feature adalah data predictor, atau dalam dunia statistik disebut sebagai variabel independen, yaitu data variabel yang dapat memprediksi target atau hasil.
Data target dikenal juga sebagai label dan dalam dunia statistik sebagai variabel dependen, yaitu data variabel yang diprediksi atau variabel yang sedang diuji dan diukur dalam percobaan, dan bergantung pada variabel independen.

- Apa Perbedaan Supervised Learning dan Unsupervised Learning?

Supervised learning merupakan pembelajaran mesin yang bisa menerima informasi yang sudah ada pada data dengan memberikan label tertentu. Supervised learning memiliki colum/data label, outputnya jelas. 

Unsupervised learning merupakan pembelajaran mesin yang digunakan pada data yang tidak memiliki informasi yang bisa diterapkan secara langsung. Unsupervised learning tidak mempunyai label, atau tidak mempunyai output yang jelas. Jadi setelah diproses, data mentah digrupkan dan kita mengenali bahwa ada pola tertentu yang akan dikenali. Kemudian kita bisa menggrupkan data tersebut berdasarkan feature atau atribut yang dimiliki. Dengan unsupervised learning ini bisa membantu kita menemukan struktur atau pola tersembunyi pada data yang tidak memiliki label.

- Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya!

Supervised learning dibagi menjadi dua, yaitu klasifikasi dan regresi. Hal yang membedakan adalah outputnya.

Output klasifikasi adalah class label, sedangkan output regresi adalah quantity yang continuous.

Contohnya kalo klasifikasi itu kita memprediksi ini hewan apa, sedangkan kalo regresi itu kita memprediksi harga sebuah barang.

- Apa perbedaan Hyperparameter dan Parameter?

Hyperparameter adalah bagian dari parameter. 

Hyperparameter didefinisikan di awal pembuatan object model, sedangkan parameter dihasilkan di akhir. Hyperparameter tidak ada sangkut pautnya dengan model, sedangkan parameter ada. Hyperparameter tidak menempel di model, tetapi didefinisikan untuk mengoptimisasi model, sedangkan paramater itu adalah sesuatu yang berada dalam model.

Contoh hyperparameter adalah fit intercept, normalize, dll.

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

dtr = DecisionTreeRegressor(max_depth=4, min_impurity_decrease=0.0, min_impurity_split=None, 
                            random_state=None, splitter='random')

x_train = x.reshape(-1,1)
y_train = y
x_test = x_new.reshape(-1,1)

dtr.fit(x_train, y_train)
y_test = dtr.predict(x_test)
y_predict = dtr.predict(x_train)

fig = plt.subplots(figsize=(12, 8))
plt.scatter(x, y, c='b', label = 'Data Asli')
plt.scatter(x_train, y_predict, c='y', label = 'Prediksi Terhadap Data Training')
plt.plot(x_test, y_test, c='g', label = 'Prediksi Terhadap Data Baru')
plt.legend()
plt.show()
tree.plot_tree(dtr);
```
